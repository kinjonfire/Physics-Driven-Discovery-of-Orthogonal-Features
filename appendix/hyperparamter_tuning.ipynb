```python
import sys
sys.path.append("../")  

import numpy as np 
import math, torch, time
import torch.nn.functional as F
from torch.optim.lr_scheduler import MultiStepLR, StepLR, MultiplicativeLR
import torch.nn as nn
import matplotlib.pyplot as plt
import sys, os
import mpl_toolkits.mplot3d as p3d
from scipy import linalg
import utility, problem
import random
import net
torch.set_default_dtype(torch.float64)

def set_seed(seed=42):
    random.seed(seed)             # Python 内置随机数种子
    np.random.seed(seed)          # NumPy 随机数种子
    torch.manual_seed(seed)       # PyTorch CPU 随机数种子
    torch.cuda.manual_seed(seed)  # PyTorch GPU 随机数种子
    torch.cuda.manual_seed_all(seed)  # 多 GPU 训练时设定所有 GPU 相同种子
    torch.backends.cudnn.deterministic = True  # 保证 CUDA 计算的确定性
    torch.backends.cudnn.benchmark = False  # 关闭自动优化（可能影响速度）

# 设置随机种子
set_seed(42)
```

```python
def u_exact(x):
    return torch.sin(3*torch.pi*x+3*torch.pi/20)*torch.cos(2*torch.pi*x+torch.pi/10)+2

def f(x):
    return -13*torch.pi**2*torch.sin(3*torch.pi*x+3*torch.pi/20)*torch.cos(2*torch.pi*x+torch.pi/10) - 12*torch.pi**2*torch.cos(3*torch.pi*x+3*torch.pi/20)*torch.sin(2*torch.pi*x+torch.pi/10) - 10*u_exact(x)

widths = [50, 100, 200, 300]
lambdas = [0.01, 0.1, 1, 10, 100]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
Pro = problem.Helmholtz(f, u_exact, u_exact, -1, 1, wavenumber=10)

for i in range(len(widths)):
    for j in range(len(lambdas)):
        print("when orthgonality regularization = %s and width = %s"%(lambdas[j], widths[i]))
        print("FNN")
        params = {"d": 1, "bodyBatch": 1000, "bdryBatch":2, "lr": 0.001, "width": 100, "depth":2, "trainStep": 1000, "writeStep": 10000, "plotStep": 10000, "decay": 0.0001, "epsilon": 1e-3, "orthogonalpenalty": 1}
        params["width"] = widths[i]
        params["orthogonalpenalty"] = lambdas[j]
        startTime = time.time()
        model_fnn = net.PD_OFM(params,connection="direct").to(device)
        optimizer = torch.optim.Adam(model_fnn.parameters(),lr=params["lr"],weight_decay=params["decay"])
        error_l2_fnn, error_l2_rel_fnn = utility.trainnew(model_fnn,Pro,device,params,optimizer)
        
        print("ResNet")
        model_res = net.PD_OFM(params,connection="resnet").to(device)
        optimizer = torch.optim.Adam(model_res.parameters(),lr=params["lr"],weight_decay=params["decay"])
        error_l2_res, error_l2_rel_res = utility.trainnew(model_res,Pro,device,params,optimizer)
        
        # print("Error of the fnn model with width %s and lambda %s is %s"%(params["width"], params["orthogonalpenalty"], error_l2_fnn))
        # print("Error of the resnet model with width %s and lambda %s is %s"%(params["width"], params["orthogonalpenalty"], error_l2_res))
        print("----------------------------------------------------------------------")
        
         
```
